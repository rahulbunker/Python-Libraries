### Why Data Visualization ?

* Data Visualization is quick,easy way to convey concepts in a universal manner.

* Data visualization is the graphical representation of information and data.

➤ visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.


## What is Matplotlib ?

➤ Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

* Matplotlib is a python library used for Data Visualization.



* Matplotlib is 2D and 3D plotting python library.

* It was introduced by john hunter in the year 2002.

```python
import numpy as np
x=["python","c","java","c++"]
y=[90,65,85,70]
z=[2,55,85,25]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.")
plt.title("puplartiy of programming ",fontsize=10)
 
plt.bar(p,y,width=0.2,color="r",label="Regex")
plt.bar(p1,z,width=0.2,color="g",label="iit delhi")

plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "Java", "C++"]
y = [90, 65, 85, 70]
z = [2, 55, 85, 25]

# Create numeric positions for the bars
x_pos = np.arange(len(x))

plt.xlabel("Language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Programming Languages", fontsize=12)

# Adjust bar positions to avoid overlap
plt.bar(x_pos - 0.1, y, width=0.2, color="r", label="Regex")
plt.bar(x_pos + 0.1, z, width=0.2, color="g", label="IIT Delhi")

# Update x-ticks to match the languages
plt.xticks(x_pos, x)

plt.legend()
plt.show()
```

```python
import numpy as np
x=["python","c","java","c++"]
y=[90,65,85,70]
z=[2,55,85,25]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.")
plt.title("puplartiy of programming ",fontsize=10)
 
plt.bar(p,y,width=0.2,color="r",label="Regex")
plt.bar(p1,z,width=0.2,color="g",label="iit delhi")

plt.xticks(p+width/2,x,rotation=60)
plt.legend()
plt.show()
```
