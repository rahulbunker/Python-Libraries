# Fill_Between Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,9,3]
plt.plot(x,y,color="g")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,9,3]
plt.plot(x,y,color="g")

plt.fill_between(x,y)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,9,3]
plt.plot(x,y,color="g")

plt.fill_between(x=[2,4],y1=2,y2=5) # start and ending points 

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,3,5,9,3])


plt.plot(x,y,color="g")

plt.fill_between(x,y,color="m",where=(x>=2) & (x<=4),alpha=0.2) # apna area jo choose kre haai bo

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
