# Area Plot

```python
x=[1,2,3,4]
area=[3,2,5,3]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4]
area=[5,2,5,3]
area1=[3,2,5,3]
area2=[3,2,10,3]

plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
x=[1,2,3,4]
area=[5,2,5,3]
area1=[3,2,5,3]
area2=[3,2,10,3]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"])
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4]
area=[5,2,5,3]
area1=[3,2,5,3]
area2=[3,2,10,3]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="zero")
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4]
area=[5,2,5,3]
area1=[3,2,5,3]
area2=[3,2,10,3]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="sym")
plt.title("python")
plt.xlabel("regex")
plt.ylabel("no")
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4]
area=[5,2,5,3]
area1=[3,2,5,3]
area2=[3,2,10,3]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="wiggle")
plt.legend(loc=2)
plt.show()
```
