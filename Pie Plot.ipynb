```python
 import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
plt.pie(x)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%")
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True)
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",radius=2)
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,labeldistance=2.3)
plt.show()

# esme names dur chle gye hai
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,labeldistance=0.2)
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,startangle=90)
plt.show() 

# kis angle / degree se chart strat krna hai
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,textprops={"fontsize":25})
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,counterclock=False  )
plt.show() 

# esme sare chart ulte ho gye mean anti clock wise so honge sare chart
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,wedgeprops={"linewidth":8,"edgecolor":"m"})
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,wedgeprops={"linewidth":8,"edgecolor":"m"},rotatelabels=True)
plt.show()  
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True)
plt.title("coder choise programming")
plt.legend()
plt.show() 
```

```python
x=[10,20,30,40]
y=["c","python","c++","java"]
ex=[0.3,0.0,0.0,0.2]
c=["y","g","b","r"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True)
plt.title("coder choise programming")
plt.legend(loc=2)
plt.show() 
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","python","c++","java"]

c=["y","g","b","r"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.title("coder choise programming")

plt.show() 
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","python","c++","java"]

c=["y","g","b","r"]

plt.pie(x,labels=y,radius=1.5,colors=c)
plt.pie(x1,radius=1,colors=c)
plt.title("coder choise programming")

plt.show() 
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","python","c++","java"]

c=["y","g","b","r"]

plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.title("coder choise programming")

plt.show() 
```
