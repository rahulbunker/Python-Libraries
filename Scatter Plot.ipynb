# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
plt.scatter(day,no)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
plt.scatter(day,no,color="r")
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
plt.scatter(day,no,color=c)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
plt.scatter(day,no,color=c,s=sizes)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
plt.scatter(day,no,color=c,s=sizes,alpha=0.2)  # dots ki transfrancy change kr skte hai
plt.title("scatter plot",fontsize=15)          # alpha=> (0 to 1)
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
plt.scatter(day,no,color=c,s=sizes,alpha=0.2,marker="*")  
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
plt.scatter(day,no,color=c,s=sizes,marker="*",edgecolor="g")  
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
plt.scatter(day,no,color=c,s=sizes,marker="*",edgecolor="g",linewidth=2)  
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
color=[10,56,96,36,300,85]
plt.scatter(day,no,color=c,s=sizes,cmap="viridis") 
plt.colorbar()
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
color=[10,56,96,36,300,85]
plt.scatter(day,no,color=c,s=sizes,cmap="Accent") # cmap ham colors ka set lene ko krte h

t=plt.colorbar()
t.set_label("color bar")
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```

```python
day=[5,2,3,5,9,1]
no=[4,6,9,2,3,7]
no2=[5,6,56,25,6,4]
c=["r","y","b","r","g","b"]
sizes=[100,2,5,60,9,300]
color=[10,56,96,36,300,85]
plt.scatter(day,no,color=c,s=sizes,cmap="Accent") 
plt.scatter(day,no2,color="r",s=sizes,cmap="Accent") 
t=plt.colorbar()
t.set_label("color bar")
plt.title("scatter plot",fontsize=15)         
plt.xlabel("day",fontsize=15)
plt.ylabel("no.",fontsize=15)

plt.show()
```
