```python
import matplotlib.pyplot as plt

num = [56, 30, 58, 10, 27, 35, 42, 12, 39, 47, 48, 39, 52, 49, 44, 55, 45, 21, 19, 35, 11, 28, 14, 47,
       11, 56, 12, 36, 13, 32, 32, 14, 33, 27, 28, 21, 40, 59, 44, 41, 40, 39, 27, 23, 35, 25, 17, 36,
       55, 12]

l = [10, 20, 30, 40, 50, 60, 70]

plt.hist(num, color="b", edgecolor="g", bins=l, label="programming")
plt.xlabel("no")
plt.ylabel("python")
plt.title("regex")

# Adding vertical and horizontal lines
plt.axvline(45, color="y", label="vertical line")
plt.axhline(10, color="r", label="horizontal line")  # Horizontal line at y=10

plt.legend()
plt.show()
```
