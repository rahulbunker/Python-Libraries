# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,3,2,5,4]
plt.step(x,y,marker="o",color="r",ms=20) # ms=size
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,3,2,5,4]
plt.step(x,y,marker="o",color="r",ms=20,mfc="g",label="python") # ms=size
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend(loc=4)
plt.show()
```
