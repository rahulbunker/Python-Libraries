# Matplotlib Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
# x=[]
# y=[]
# plt.stem(x,y)
# plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y)
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="r*")
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="ro")
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=5)
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g")
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python")
plt.legend()
plt.show()
```

```python
x=[2,5,6,3,4]
y=[2,5,4,9,8]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show()
```
